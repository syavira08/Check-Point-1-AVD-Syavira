### **Outliers Values**

Dari hasil pengecekan nilai ekstrem (outlier) pada kolom bertipe numerik berikut, dapat terlihat kolom:
*   **product_id**

    Dengan presentase sekitar 12%

*   **beauty_point_earned**

    Dengan presentase sekitar 4%

*   **average_rating**

    Dengan presentase sekitar 24%

*   **total_reviews**

    Dengan presentase sekitar 15%

*   **total_recommended_count**

    Dengan presentase sekitar 15%

*   **total_repurchase_maybe_count**

    Dengan presentase sekitar 15%

*   **total_repurchase_no_count**

    Dengan presentase sekitar 14%

*   **total_repurchase_yes_count**

    Dengan presentase sekitar 15%

*   **total_in_wishlist**

    Dengan presentase sekitar 11%    

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df)
```

## **Outliers**

Pada bagian ini, kita akan menangani kolom-kolom yang memiliki outlier atau nilai ekstrem, dengan catatan sebagai berikut:
*   Jika data outliers ada <= 10% dari banyaknya
jumlah data maka dapat dihapus, tetapi jika >= 10% bisa diabaikan (tetapi tetap melihat
konteks dari kolom yang memiliki outliers) hal
ini tidak bisa menjadi acuan utama.
*   Alternatif lain dapat melakukan imputasi
dengan metode capping dan flooring.

### **1. product_id (12%)**

Kolom product_id memiliki persentase outlier sebesar 12%, karena kolom ini berfungsi sebagai identitas unik setiap produk, nilai di dalamnya tidak dapat diperlakukan sebagai data numerik biasa untuk dianalisis outlier. Selain itu, struktur ID produk yang bersifat berurutan dan mencakup lebih dari 7.500 produk sudah sesuai dengan konteks dataset. Oleh karena itu, kolom ini sebaiknya diabaikan dalam penanganan outliers.

### **2. beauty_point_earned (4%)**

Untuk kolom beauty_point_earned yang memiliki outlier sebesar 4%, saya memutuskan untuk menghapusnya. Persentase outlier yang relatif kecil serta fungsi kolom ini yang hanya berisi jumlah beauty points yang diperoleh pembeli membuatnya kurang relevan untuk analisis popularitas brand atau produk. Karena tidak memberikan kontribusi penting terhadap tujuan analisis, kolom ini dapat dihapus tanpa memengaruhi hasil akhir.

### **3. average_rating (24%)**

Kolom average_rating tidak diabaikan dan tetap dilakukan imputasi karena kolom ini berisi penilaian pelanggan dengan rentang nilai 1 sampai 5. Nilai rating tidak seharusnya berada di luar batas tersebut, sehingga setiap nilai ekstrem perlu diperbaiki agar tetap sesuai dengan karakteristik data aslinya.

```python
columns_to_impute = ["average_rating"]

for col in columns_to_impute:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    df.loc[:, col] = df[col].clip(lower=lower_bound, upper=upper_bound)
```

### **4. total_reviews (15%)**

Kolom total_reviews yang memiliki presentase 15% lebih baik di abaikan karena kolom ini mencakup data total ulasan yang diberikan oleh pelanggan, yang dimana jumlah review yang diberikan pelanggan itu tidak dapat dikendalikan dan tergantung dengan popularitas produknya juga, jadi nilai ekstrem pada kolom ini tidak dapat dianggap sebagai outlier yang harus diperbaiki.

### **5. total_recommended_count (15%)**

Untuk kolom total_recommended_count dengan presentase 15% juga lebih baik di abaikan, karena sama hal nya dengan kolom total_reviews yang dimana data rekomendasi yang diberikan oleh pelanggan itu tidak bisa dikendalikan dan tergantung dengan popularitas produknya juga.

### **6. total_repurchase_maybe_count (15%), total_repurchase_no_count (14%), dan total_repurchase_yes_count (15%)**

Untuk ketiga kolom tersebut memiliki konteks yang sama juga dengan kolom sebelumnya yaitu total_recommended_count dan total_reviews, yang dimana mencakup nilai data yang berasal dari pelanggan dan bergantung dengan popularitas produk, sehingga nilai ekstrem pada kolom tidak dapat dianggap outliers dan lebih baik diabaikan.

### **9. total_in_wishlist (11%)**

Untuk kolom ini juga mengandung nilai yang berasal dari pelanggan yang juga bergantung pada popularitas produk sehingga nilai ekstrem tidak dapat dianggap outliers dan dapat diabaikan.

Selanjutnya bisa dilakuakan pengecekan ulang untuk memastikan bahwa outliers yang kita imputasi tadi sudah tidak mengandung ouliers lagi.

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df)
```
